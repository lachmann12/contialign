```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
kmer22 = pd.read_csv("../output/transcript_info.tsv", sep="\t", header=None)
kmer22.index = kmer22.iloc[:,0]
kmer31 = pd.read_csv("../output/transcript_info_31.tsv", sep="\t", header=None)
kmer31.index = kmer31.iloc[:,0]

kmer40 = pd.read_csv("../output/transcript_info_40.tsv", sep="\t", header=None)
kmer40.index = kmer40.iloc[:,0]

idx = set(kmer22.iloc[:,0]).intersection(set(kmer31.iloc[:,0]))
idx = idx.intersection(set(kmer40.iloc[:,0]))

kmer22 = kmer22.loc[idx,:]
kmer31 = kmer31.loc[idx,:]
kmer40 = kmer40.loc[idx,:]

kmer22_d = kmer22.iloc[:,1]/kmer22.iloc[:,2]
kmer31_d = kmer31.iloc[:,1]/kmer31.iloc[:,2]
kmer40_d = kmer40.iloc[:,1]/kmer40.iloc[:,2]
```

```python
plt.scatter(kmer22_d, kmer40_d, s=0.2)
plt.xlabel("k-mer length 22")
plt.ylabel("k-mer length 40")
```

```python
pd.DataFrame(np.array([kmer22_d, kmer31_d, kmer40_d]).T, columns=["22","31","40"]).plot.density(figsize = (7, 4), linewidth = 4, xlim=(-0.2,1.2))
```
